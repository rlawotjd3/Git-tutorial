```python
menu = {
    "햄": 5800,
    "써브웨이 클럽": 7100,
    "에그마요": 5500,
    "이탈리안 비엠티": 6700,
    "폴드포크": 7200,
    "쉬림프": 7100
}

selected_menu_list = []
admin_password = "0000"

def order():
    print("----- 메뉴 -----")
    for item, price in menu.items():
        print(item + ": " + str(price) + "원")

    selected_menu = input("메뉴를 선택하세요: ")
    while selected_menu not in menu:
        print("잘못된 메뉴입니다. 다시 입력해주세요.")
        selected_menu = input("메뉴를 선택하세요: ")

    length = input("길이를 선택하세요(15cm 또는 30cm): ")
    while length not in ["15cm", "30cm"]:
        print("잘못된 길이입니다. 다시 입력해주세요.")
        length = input("길이를 선택하세요(15cm 또는 30cm): ")

    price = menu[selected_menu]
    if length == "30cm":
        price += 5500

    set_choice = input("세트로 하시겠습니까? (예/아니오): ")
    while set_choice not in ["예", "아니오"]:
        print("잘못된 입력입니다. 다시 입력해주세요.")
        set_choice = input("세트로 하시겠습니까? (예/아니오): ")

    if set_choice == "예":
        price += 2600

    bread = input("빵 종류를 선택하세요(허니오트, 하티, 위트, 파마산 오레가노, 화이트, 플랫브래드): ")

    vegetables = input("제외할 야채 종류를 입력하세요(콤마로 구분): ").split(",")

    cheese = input("치즈 종류를 선택하세요(아메리칸 치즈, 슈레드 치즈, 모짜렐라 치즈): ")

    sauces = []
    for i in range(2):
        sauce = input(f"{i+1}번째 소스 종류를 선택하세요(랜치, 마요네즈, 스위트 어니언, 허니머스타드, 스위트칠리, 바비큐, 머스타드, 스모크 바비큐, 소금, 후추): ")
        sauces.append(sauce)

    selected_menu_list.append(selected_menu)

    print("----- 합계 금액 -----")
    print("선택한 메뉴:", ', '.join(selected_menu_list))
    print("가격:", price, "원")

    payment = int(input("입금 금액을 입력하세요: "))

    balance = payment - price
    if balance < 0:
        print("입금 금액이 부족합니다.")
        return

    if balance >= 0:
        print("잔돈:", balance, "원 입니다. 이용해주셔서 감사합니다.")

    print("----- 영수증 -----")
    print("선택한 메뉴:", ', '.join(selected_menu_list))
    print("빵 종류:", bread)
    print("제외한 야채 종류:", ', '.join(vegetables))
    print("치즈 종류:", cheese)
    print("소스 종류:", ', '.join(sauces))
    print("가격:", price, "원")
    print("입금 금액:", payment, "원")
    if balance >= 0:
        print("잔돈:", balance, "원")
    return    

def admin_mode():
    global admin_password

    password = input("비밀번호를 입력하세요: ")
    while password != admin_password:
        print("잘못된 비밀번호입니다.")
        password = input("비밀번호를 입력하세요: ")

    while True:
        print("----- 관리자 모드 -----")
        print("1. 메뉴 추가")
        print("2. 메뉴 삭제")
        print("3. 관리자 모드 종료")
        print("4. 비밀번호 변경")
        print("5. 프로그램 종료")

        choice = input("기능을 선택하세요: ")

        if choice == "1":
            item = input("추가할 메뉴를 입력하세요: ")
            price = int(input("메뉴의 가격을 입력하세요: "))
            menu[item] = price
            print("메뉴가 추가되었습니다.")

        elif choice == "2":
            item = input("삭제할 메뉴를 입력하세요: ")
            if item in menu:
                del menu[item]
                print("메뉴가 삭제되었습니다.")
            else:
                print("존재하지 않는 메뉴입니다.")

        elif choice == "3":
            print("관리자 모드를 종료합니다.")
            break

        elif choice == "4":
            current_password = input("현재 비밀번호를 입력하세요: ")
            if current_password != admin_password:
                print("비밀번호가 일치하지 않습니다.")
                continue

            new_password = input("새로운 비밀번호를 입력하세요: ")
            confirm_password = input("새로운 비밀번호를 한 번 더 입력하세요: ")

            while new_password != confirm_password:
                print("비밀번호가 일치하지 않습니다. 다시 입력해주세요.")
                new_password = input("새로운 비밀번호를 입력하세요: ")
                confirm_password = input("새로운 비밀번호를 한 번 더 입력하세요: ")

            admin_password = new_password
            print("비밀번호가 변경되었습니다.")

        elif choice == "5":
            print("프로그램을 종료합니다.")
            exit()
            
        else:
            print("잘못된 선택입니다. 다시 선택해주세요.")
            
            
def main():
    while True:
        print("----- 서브웨이 키오스크 ----")
        print("1. 주문하기")
        print("2. 관리자 기능")
        choice = input("원하는 기능을 선택하세요: ")

        if choice == "1":
            order()

        elif choice == "2":
            admin_mode()

        else:
            print("잘못된 선택입니다. 다시 선택해주세요.")

main()
```
